```python
import pandas as pd
from IPython import display
```

```python
df = pd.read_csv('/content/HDR21-22_Composite_indices_complete_time_series.csv')
```

```python
df.columns
```

```python
df_cn_usa = df[df['country'].isin(['China', 'United States'])]
```

```python
df_melt = pd.melt(df_cn_usa, id_vars=['country', 'iso3', 'hdicode', 'region'])
```

```python
df_melt['variable_name'] = df_melt.variable.str.split('_').apply(lambda x: "_".join(x[:len(x)-1]))
df_melt['years'] = df_melt.variable.str.split('_').apply(lambda x: x[len(x)-1])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style="darkgrid")
sns.set(rc={'figure.figsize':(13.7,4.27)})
sns.set(font_scale=0.5)


# Plot the responses for different events and regions
```

```python
#for cty in df_melt.country.unique():
for var in df_melt.variable_name.unique():
    
  
  print(var.upper())
  print(".....................")
  aux = df_melt.loc[ (df_melt.variable_name == var), ['years','value', 'country']].sort_values('years')
  sns.lineplot(x="years", y="value",
            hue="country",
            data=aux)
  plt.show()
  
```
